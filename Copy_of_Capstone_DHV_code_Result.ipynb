```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#DATA COLLECTION
# Read the CSV file into a DataFrame
df = pd.read_csv('/content/sales.csv',encoding="latin1")
```

```python
#DATA CLEANING

# Display the first few rows of the DataFrame
print("Initial dataset:")
print(df.head())
```

```python
#DATA CLEANING
# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Handle missing values (if any)
# For example, you can drop rows with missing values
# df.dropna(inplace=True)

# Check for duplicate rows
print("\nDuplicate rows:")
print(df.duplicated().sum())

# Handle duplicate rows (if any)
# For example, you can drop duplicate rows
# df.drop_duplicates(inplace=True)

# Check for data types of columns
print("\nData types of columns:")
print(df.dtypes)
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
# DATA PRE PROCESSING
# Save the cleaned dataset to a new CSV file
# df.to_csv('cleaned_dataset.csv', index=False
```

```python
# DATA PRE PROCESSING
print("data pre processing")
print("\nCleaned dataset:")
print(df.head())

# Save the cleaned dataset to a new CSV file
# df.to_csv('cleaned_dataset.csv', index=False)
```

```python
#DATA VISUALIZATION


# Bar plot: Count of transactions per country
plt.figure(figsize=(10, 6))
sns.countplot(x='Country', data=df, hue='Country', palette='pastel', legend=False)
plt.title('Count of Transactions per Country')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Histogram: Distribution of prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Line plot: Quantity over time
plt.figure(figsize=(10, 6))
df['Date'] = pd.to_datetime(df['Date'])  # Convert 'Date' column to datetime format
quantity_over_time = df.groupby('Date')['Quantity'].sum().reset_index()
sns.lineplot(x='Date', y='Quantity', data=quantity_over_time, marker='o', color='green')
plt.title('Quantity Over Time')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
